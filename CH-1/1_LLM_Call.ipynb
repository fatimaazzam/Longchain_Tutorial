# **LLM CALL**


```python
from langchain_openai import ChatOpenAI
import os
import load_dotenv

# This function will load all the variables from the .env file and 
# make them available in the os.environ dictionary (env variables)
load_dotenv.load_dotenv() 

if os.environ.get("OPENAI_API_KEY"):
    print("API veriable exists")
else:
    raise ValueError("OPENAI_API_KEY not found")
```

```python
llm_openai = ChatOpenAI(
    model = "gpt-5-mini",
    temperature = 0
)

llm_openai.invoke("Bro, Tell fun fact")
```

```python
from langchain.chat_models import  init_chat_model

llm_openai = init_chat_model(model = "gpt-5-mini")
llm_openai.invoke("How are you")
```
