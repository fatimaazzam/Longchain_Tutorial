# **Messages**

```python
from langchain_openai import ChatOpenAI
import os
import load_dotenv

# This function will load all the variables from the .env file and 
# make them available in the os.environ dictionary (env variables)
load_dotenv.load_dotenv() 

if os.environ.get("OPENAI_API_KEY"):
    print("API veriable exists")
else:
    raise ValueError("OPENAI_API_KEY not found")
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

llm_openai = ChatOpenAI(
    model = "gpt-5-mini",
    temperature = 0
)


my_messages = [
    SystemMessage(content="You are gen-z assistence who always answer in a fun way"),
    HumanMessage(content="Tell me fun fact")
]
llm_openai.invoke(my_messages).content
```

# **Prompts**
Prompts are the messages that are sent to the LLM. Prompts are more user friendly than messages


```python
my_messages = [
    SystemMessage(content="You are gen-z assistence who always answer in a fun way"),
    HumanMessage(content="Tell me fun fact about Honey")
]

llm_openai.invoke(my_messages).content
```

```python
from langchain_core.prompts import PromptTemplate


user_input = input("Enter a topic for fun fact: ")

dynamic_prompt = PromptTemplate.from_template("Write a fun about {topic}")

ready_prompt = dynamic_prompt.invoke({"topic": user_input})

llm_openai.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import  ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {tone} assistant"),
    ("user", "Write a fun fact about {topic}")
    
])

user_input = input("Enter a tpoic: ")
tone_input = input("Enter a tone: ")
ready_prompt = prompt_template.invoke({"topic": user_input, "tone": tone_input})
llm_openai.invoke(ready_prompt.messages).content
```
